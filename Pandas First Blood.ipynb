```python
# загружаем файл с рейтингом
data = pd.read_csv(r'C:\Users\admin1\Desktop\ДЗ Мамы\DZ 10 Pandas Первая кровь\DZ 10 Pandas\ml-latest-small\ratings.csv')
data.head()
```

```python
# загружаем файл с названиями фильмов и жанрами

nameMovies = pd.read_csv(r'C:\Users\admin1\Desktop\ДЗ Мамы\DZ 10 Pandas Первая кровь\DZ 10 Pandas\ml-latest-small\movies.csv')
nameMovies.head()
```

**Version.1**

```python
короткий способ найти нужно id: задаем условие, что выбираем только те строчки, где рейтинг равен 5 
и создаем серию из столбца айди фильма. После чего производим подсчет каждого id в серии

count5 = data[data.rating == 5]['movieId'].value_counts()
count5.head()
```

```python
# Попытка слить с таблицей названий фильмов Здесь выяснилось, что 'Series' object has no attribute 'merge'.
# какие есть альтернативы сделать соединение для объектов типа series?

count5.merge(nameMovies, how='left') 
```

```python
# Откатываем на шаг назад и делаем не серию, а таблицу из двух столбцов. Запускаем и получаем 
# новую ошибку связанную с отсутствием подходящих столбцов для проведения слияния. В урезанной табличке 
# уже новые названия столбцов

count5v1 = data[data.rating == 5]['movieId'].value_counts().reset_index(name='count')

# переименовываем столбцы для последующего мерджа по movieId
count5v1.columns = ['movieId', 'count5reit']
count5v1.head()
```

```python
count5v1.merge(nameMovies, how='left').head()
```

#### Exercise 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики 
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года.
Не учитывайте в расчетах отрицательные значения quantity.


```python
# Читаем файл

power = pd.read_csv('power.csv')
power.head()
```

```python
# Отбираем множественными условиями нужный сет - сначала отбираем по трем условиям.

power_prepared = power[power['country'].isin(["Latvia", "Lithuania", "Estonia"]) 
      & power['category'].isin([4,12,21])
      & ((power['year'] >= 2005) 
      & (power['year'] <= 2010))
      & (power_prepared["quantity"] >= 0)]

power_prepared
```

```python
# Сначала были убраны все строки, что строго больше нуля. Но быыло замечено расхождение в количестве 
# итоговых строк при разных запросах. Проверка показала, что нулевые значения в таблице тоже 
# присутствуют. По условию задачи говориться "Не учитывайте в расчетах отрицательные значения quantity.". 
# Поэтому было решено оставить нулевые. Убираем только отрицательные (!) значения "quantity"

power_prepared[(power_prepared["quantity"] == 0)]
```

```python
#  суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
# категорий 4, 12 и 21 за период с 2005 по 2010 года
power_prepared["quantity"].sum()
```
